### 如何导入matplotlib，pylab，pyplot，numpy等

子模块“ matplotlib.pyplot”为绘图库提供了一个面向对象的接口。 matplotlib文档中的许多示例都遵循导入约定，将“ matplotlib.pyplot”导入为“ plt”，将“ numpy”导入为“ np”。当然，完全由用户决定是否以名称np（在matplotlib示例中经常这样做）或以文本中的N（在早期numpy的前期）下导入numpy库。被称为“数字”）或您喜欢的任何其他名称。类似地，是否将绘图子模块（`matplotlib.pyplot`）导入为`plt`，就像在matplotlib文档或`plot`（可能会说得更清晰）一样，这也是一个问题。

与往常一样，必须在个人喜好和选择这些名字的惯常做法之间取得平衡。如果该代码很可能会被他人使用或发布，则与通用用法保持一致当然更为重要。

绘制几乎总是需要数字数据数组，因此，`numpy`模块被大量使用：它为Python提供了快速且内存高效的数组处理（请参阅第14章（14-numpy.ipynb））。

因此，我们也可以像上面的示例1b一样编写上面的示例1a（其功能与上面的示例相同，并将创建相同的图）：

````python
import pylab
import numpy as N

x = N.arange (-3.14, 3.14, 0.01)
y = N.sin(x)

pylab.plot(x, y)
````

因为numpy.arange和numpy.sin对象已经被导入到pylab名称空间中，所以我们也可以将其编写为示例1c：

````python
#example 1c
import pylab as p

x = p.arange(-3.14, 3.14, 0.01)
y = p.sin(x)

p.plot(x, y)
````

如果我们真的想减少输入的字符，我们也可以从pylab便捷模块中获取全部功能，并将代码重写为示例1d：

````python
#example 1 d
from pylab import *  # not generally recommended
                     # okay for interactive testing

x = arange(-3.14, 3.14, 0.01)
y = sin(x)
plot(x, y)
show()
````

这可能是非常方便的，但带有严重的健康警告：

### 微调您的情节

### 可视化矩阵数据

下面的程序演示了如何创建矩阵条目的位图图。

````python
import numpy as np
import matplotlib.pyplot as plt

# Helper function (from https://github.com/matplotlib/matplotlib/blob/master/lib/matplotlib/mlab.py
# as of August 2018)
def bivariate_normal(X, Y, sigmax=1.0, sigmay=1.0,
                     mux=0.0, muy=0.0, sigmaxy=0.0):
    """
    Bivariate Gaussian distribution for equal shape *X*, *Y*.
    See `bivariate normal
    <http://mathworld.wolfram.com/BivariateNormalDistribution.html>`_
    at mathworld.
    """
    Xmu = X-mux
    Ymu = Y-muy

    rho = sigmaxy/(sigmax*sigmay)
    z = Xmu**2/sigmax**2 + Ymu**2/sigmay**2 - 2*rho*Xmu*Ymu/(sigmax*sigmay)
    denom = 2*np.pi*sigmax*sigmay*np.sqrt(1-rho**2)
    return np.exp(-z/(2*(1-rho**2))) / denom

# create matrix Z that contains some interesting data
delta = 0.1
x = y = np.arange(-3.0, 3.0, delta)
X, Y = np.meshgrid(x, y)
Z = bivariate_normal(X, Y, 3.0, 1.0, 0.0, 0.0)

# display the 'raw' matrix data of Z in one figure
plt.figure(1)
plt.imshow(Z, interpolation='nearest')
plt.title("imshow example 1a: no interpolation")
plt.savefig("pylabimshow1a.pdf")

# display the data interpolated in other figure
plt.figure(2)
im = plt.imshow(Z, interpolation='bilinear')
plt.title("imshow example 1b: with bi-linear interpolation")
plt.savefig("pylabimshow1b.pdf")
````

为了使用不同的颜色图，我们使用`matplotlib.cm`模块（其中cm代表颜色图）。以下代码演示了如何从已提供的地图集中选择颜色图，以及如何修改颜色图（此处通过减少地图中的颜色数量）。最后一个示例模仿了matplotlib随附的更复杂的“ contour”命令的行为。

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm        # Colour map submodule

#create matrix Z that contains some data interesting data
delta = 0.025
x = y = np.arange(-3.0, 3.0, delta)
X, Y = np.meshgrid(x, y)
Z = bivariate_normal(X, Y, 3.0, 1.0, 0.0, 0.0)

Nx, Ny = 2, 3
plt.subplot(Nx, Ny, 1)  # next plot will be shown in
                        # first subplot in Nx x Ny
                        # matrix of subplots

plt.imshow(Z, cmap=cm.jet)  # default colourmap 'jet'
plt.title("colourmap jet")

plt.subplot(Nx, Ny, 2)   # next plot for second subplot
plt.imshow(Z, cmap=cm.jet_r)  # reverse colours in jet
plt.title("colourmap jet_r")

plt.subplot(Nx, Ny, 3)
plt.imshow(Z, cmap=cm.gray)
plt.title("colourmap gray")

plt.subplot(Nx, Ny, 4)
plt.imshow(Z, cmap=cm.hsv)
plt.title("colourmap hsv")

plt.subplot(Nx, Ny, 5)
plt.imshow(Z, cmap=cm.gist_earth)
plt.title("colourmap gist_earth")

plt.subplot(Nx, Ny, 6)
# make isolines by reducing number of colours to 10
mycmap = cm.get_cmap('jet', 10)    # 10 discrete colors
plt.imshow(Z, cmap=mycmap)
plt.title("colourmap jet\n(10 colours only)")

plt.savefig("pylabimshowcm.pdf")
````

### 设置动画的帧频

Visual Python的特别优势在于它能够显示与时间有关的数据：

- 一个非常有用的命令是`rate()`命令，该命令可确保仅以一定的帧速率执行循环。这是一个示例，每秒精确打印两个“ Hello World”（`vpythondemo2.py`）：

```python
import visual

for i in range(10):
    visual.rate(2)
    print("Hello World (0.5 seconds per line)")
```

- 所有Visual Python对象（例如上例中的球体和盒子）都具有一个.pos属性，该属性包含位置（对象\[sphere,box\]的中心或对象\[cylinder,helix\]的一端）。这是一个显示球体上下运动的示例（`vpythondemo3.py`）：

```python
import visual, math

ball = visual.sphere()
box = visual.box( pos=[0,-1,0], width=4, length=4, height=0.5 )

#tell visual not to automatically scale the image
visual.scene.autoscale = False

for i in range(1000):
    t = i*0.1
    y = math.sin(t)

    #update the ball's position
    ball.pos = [0, y, 0]

    #ensure we have only 24 frames per second
    visual.rate(24)
```

### 跟踪轨迹

您可以使用“曲线”跟踪对象的轨迹。基本思想是将位置附加到该曲线对象，如本示例所示（`vpythondemo4.py`）：

```python
import visual, math

ball = visual.sphere()
box = visual.box( pos=[0,-1,0], width=4, length=4, height=0.5 )
trace=visual.curve( radius=0.2, color=visual.color.green)

for i in range(1000):
    t = i*0.1
    y = math.sin(t)

    #update the ball's position
    ball.pos = [t, y, 0]

    trace.append( ball.pos )

    #ensure we have only 24 frames per second
    visual.rate(24)
```

### 连接对象（圆柱体，弹簧等）

圆柱和螺旋可用于“连接”两个对象。除了“ pos”属性（用于存储对象一端的位置）外，还有“ axis”属性用于存储从“ pos”指向对象另一端的向量。这是一个显示圆柱体的示例：（`vpythondemo5py`）：

```python
import visual, math

ball1 = visual.sphere( pos = (0,0,0), radius=2 )
ball2 = visual.sphere( pos = (5,0,0), radius=2 )
connection = visual.cylinder(pos = ball1.pos, \
                             axis = ball2.pos - ball1.pos)

for t in range(100):
    #move ball2 
    ball2.pos = (-t,math.sin(t),math.cos(t))

    #keep cylinder connection between ball1 and ball2
    connection.axis = ball2.pos - ball1.pos

    visual.rate(24)
```

- `stereodepth = 1`：屏幕表面的3D场景（通常看起来最好）

可视化高维数据
-----------------------------------

通常，我们需要了解在空间3d位置定义的数据。数据本身通常是标量场（例如温度）或3d向量（例如速度或磁场），或者有时是张量。例如，对于在3d空间（$\vec{f}(\vec{x})$，其中$\vec{x} \in I\!\!R^3$和$\vec{f}(\vec{x})
\in I\!\!R^3$）中定义的3d矢量场*f*，我们可以在空间中的每个（网格）点绘制3d箭头。这些数据集通常与时间相关。

在科学与工程中最可能用来可视化此类数据集的库可能是VTK，即可视化工具包（<http://vtk.org>）。这是一个实质性的C ++库，具有与高级语言（包括Python）的接口。

可以直接从Python代码调用这些例程，也可以将数据以VTK库可以读取的格式（称为vtk数据文件）写入磁盘，然后使用Mayavi，ParaView和VisIt等独立程序进行读取这些数据文件并对其进行操作（使用GUI进行操作）。所有这三个都在内部使用VTK库，并且可以读取vtk数据文件。
